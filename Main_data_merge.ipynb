```r
# Load dplyr
library(dplyr)
library(readxl)
```

```r
wis <- read.csv("wcld.csv",, header = TRUE)
head(wis)
colnames(wis)
```

```r
county <-  read_excel("Wisconsin County-FIPS mapping.xlsx")
head(county)
```

```r
# Perform left join
merged_data <- wis %>%
  left_join(county, by = c("county" = "County Code"))

#Rename FIPS and County name columns
merged_data <- merged_data %>%
  rename(
    County_Name = `County Name`,
    FIPS = `FIPS Code`
  )

# View the result
head(merged_data)
```

```r
#Checking missing values in FIPS column
sum(is.na(merged_data$FIPS))
any(is.na(merged_data$FIPS))
any(is.na(merged_data$County_name))
```

- There are no missing values in FIPS Column or County names column

```r
#Write data
write.csv(merged_data, "main_merged_data_FIPS.csv", row.names = FALSE)
```
